```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'rx')
plt.show()
```

```python
plt.plot(x,y, 'rs-')
plt.show()
```

```python
plt.plot(x,y, 'yD:')
plt.show()
```

```python
plt.hist(x)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(x,y+2.5, 'x', markeredgewidth=2)
ax.plot(x,y,linewidth=2.0)
ax.plot(x,y-2.5,'o-',linewidth=2)

ax.set(xlim=(0, 8), xticks=np.arange(1,8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
