```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
print(plt.style.available)
```

```python
plt.style.use('bmh')
```

```python
country = ['INDIA', 'MEXICO','JAPAN','COLOMBIA','GERMANY']
population = [1000,800,900,1000,300]
```

```python
plt.bar(country,population)
plt.show()
```

```python
plt.bar(country,population, width=0.5, color=['red','blue'])
plt.xticks(np.arange(5),('India','Japon','Mexico','Colombia','Alemania'), rotation=45)
plt.title('POPULATION BY COUNTRY')
plt.ylabel('Population') 
plt.xlabel('Country') 
plt.show()
```

```python
country = ['INDIA', 'MEXICO','JAPAN','COLOMBIA','GERMANY']
population = [1000,800,900,1000,300]
```

```python
plt.bar(country,population, width=0.5, color=['red','blue'])
plt.xticks(np.arange(5),('India','Japon','Mexico','Colombia','Alemania'), rotation=45)
plt.title('POPULATION BY COUNTRY', color='gray', fontweight='bold', fontsize=20, fontname='Verdana')
plt.ylabel('Population', color='c', fontweight='bold', fontsize=16, fontname='Verdana') 
plt.xlabel('Country', color='y', fontweight='bold', fontsize=16, fontname='Verdana') 
plt.show()
```

```python
plt.barh(country,population,height=0.6)
plt.show()
```
