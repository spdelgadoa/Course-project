```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
print(plt.style.available)
```

```python
plt.style.use('Solarize_Light2')
```

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,'r-')
ax.plot(x,x+2,'b--')
ax.plot(x,x+3,'y:')
ax.plot(x,x+4,'g-.')
```

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1, color ='mediumseagreen', alpha=0.3, linewidth=5, linestyle='-', marker='o', markersize=10, markerfacecolor='white')
ax.plot(x,x+2,color ='c', linewidth=5, linestyle='--', marker='x')
ax.plot(x,x+3,color ='#e377c2', linewidth=3, linestyle='dashed', marker='v')
ax.plot(x,x+4,color ='#17becf', linewidth=2, linestyle=':', marker='P')
```
