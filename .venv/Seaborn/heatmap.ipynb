```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
numeric_tips = tips.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_tips.corr()
correlation_matrix
```

```python
sns.heatmap(correlation_matrix)
plt.show()
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap=sns.cubehelix_palette(as_cmap=True), linewidth=5, linecolor='gray',
vmin=0.5, vmax=1, cbar=True)
plt.show()
```
