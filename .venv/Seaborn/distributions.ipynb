```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
tips = sns.load_dataset('tips')

tips.head()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat="count", multiple='dodge')
plt.show()
```

```python
sns.kdeplot(data=tips,x='tip',hue='sex',cumulative=False,fill=True,bw_adjust=1)
plt.show()
```

```python
sns.ecdfplot(data=tips,x='tip',hue='sex',stat='count')
plt.show()
```

```python
sns.displot(data=tips,x='tip',hue='sex',kind='hist', multiple='stack')
plt.show()
```

RETO

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", style='time')
plt.show()
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="size")
plt.show()
```

```python
fmri = sns.load_dataset("fmri")
sns.relplot(data=fmri, x="timepoint", y="signal", kind="line")
```

```python
tips = sns.load_dataset("tips")
sns.catplot(data=tips, x="day", y="total_bill", hue='sex', palette='pastel', kind='swarm')
```
